# Matplotlib 基礎

```python
import matplotlib.pyplot as plt
import numpy as np
```

### Matplotlib document
https://matplotlib.org/
- matplotlib はグラフを書くためにパーツを当てはめて描いている
    1. Figure : グラフを書くための台紙
    2. Axes : アクセス(title, Axis[軸名])などの細かい設定ができる

```python
fig = plt.figure(figsize=(5, 5)) # figsize : 台紙のサイズ (記述しなくても default の size が入る)
ax = plt.axes()
```

### 折れ線グラフ

```python
x = [105, 210, 301, 440, 500]
y = [10, 20, 30, 50, 60]
fig = plt.figure(figsize=(5, 5))
ax = plt.axes()

ax.plot(x, y) # plot になっている所に注意
plt.show() # グラフを表示させる為
```

```python
x = [105, 210, 301, 440, 500]
y = [10, 20, 30, 50, 60]
fig = plt.figure(figsize=(5, 5))
ax = plt.axes()

ax.set_title('practice matplotlib') # title 記述方
ax.set_xlabel('x') # x 軸の名前
ax.set_ylabel('y') # y 軸の名前
ax.grid() # grid線の記述

ax.plot(x, y)
plt.show()
```

### グラフの線や色の編集
- plot(引数)：引数の中に記述していく

```python
x = [105, 210, 301, 440, 500]
y = [10, 20, 30, 50, 60]
fig = plt.figure(figsize=(5, 5))
ax = plt.axes()

ax.set_title('practice matplotlib') # title 記述方
ax.set_xlabel('x') # x 軸の名前
ax.set_ylabel('y') # y 軸の名前
ax.grid() # grid線の記述

ax.plot(x, y, marker="o", color="red", linestyle='--', linewidth = 2.0) # marker,color,style,width
plt.show()
```

### 棒グラフの記述

```python
x = ["A", "B", "C", "D"]
y = [400, 300, 200, 100]
fig = plt.figure(figsize=(5, 5))
ax = plt.axes()

ax.bar(x, y) # bar になっている所に注意(棒グラフの場合は bar)
plt.show() # グラフを表示させる為
```

### ヒストグラム
- ある特定の data を特定の期間で区切って、区間の個数、バラつきを棒グラフで表したもの

numpy を使用して乱数を発生させて、乱数を使用してヒストグラムを記述していく

```python
x = np.random.randn(1000) # 正規分布に従った乱数を発生させている：1000個のdata を発生させる

fig = plt.figure(figsize=(5, 5))
ax = plt.axes()

ax.hist(x) # hist になっている所に注意
plt.show() # グラフを表示させる為
```

```python
x = np.random.randn(1000) # 正規分布に従った乱数を発生させている：1000個のdata を発生させる

fig = plt.figure(figsize=(5, 5))
ax = plt.axes()

ax.hist(x, bins=50) # bins : 区間の数（幅）を指定
plt.show() # グラフを表示させる為
```

### 散布図
- 縦軸と横軸対して data を点で plot した形

- ある範囲での整数を発生させる

1 ~ 100 まで　ランダムの数字を　200個発生させる

```python
x = np.random.randint(1, 101, 200)
y = np.random.randint(1, 101, 200)

fig = plt.figure(figsize=(5, 5))
ax = plt.axes()

ax.scatter(x, y) # scatter と記述
plt.show() # グラフを表示させる為
```

## 複数のグラフを記述
- サブプロットというパーツを追加していく

```python
fig = plt.figure()

ax1 = fig.add_subplot(221) # ２行２列の１番目に記述　
ax2 = fig.add_subplot(222) # .......2番目
ax3 = fig.add_subplot(223) # .......3番目
ax4 = fig.add_subplot(224) # .......4番目
```

### subplot adjust : 各　pltの間隔を調節

```python
fig = plt.figure()

ax1 = fig.add_subplot(221) # ２行２列の１番目に記述　
ax2 = fig.add_subplot(222) # .......2番目
ax3 = fig.add_subplot(223) # .......3番目
ax4 = fig.add_subplot(224) # .......4番目

plt.subplots_adjust(wspace=0.5, hspace=0.5)
```

```python
fig = plt.figure(figsize=(10, 10))

ax1 = fig.add_subplot(221) # ２行２列の１番目に記述　
ax2 = fig.add_subplot(222) # .......2番目
ax3 = fig.add_subplot(223) # .......3番目
ax4 = fig.add_subplot(224) # .......4番目

plt.subplots_adjust(wspace=0.5, hspace=0.5)

# ax1
ax1.set_title('ax1')
x = [105, 210, 301, 440, 500]
y = [10, 20, 30, 50, 60]
ax1.plot(x, y) # plot になっている所に注意

# ax2
ax2.set_title('ax2')
x = ["A", "B", "C", "D"]
y = [400, 300, 200, 100]
ax2.bar(x, y) # bar になっている所に注意(棒グラフの場合は bar)

# ax3
ax3.set_title('ax3')
x = np.random.randn(1000) # 正規分布に従った乱数を発生させている：1000個のdata を発生させる
ax3.hist(x) # hist になっている所に注意

# ax4
ax4.set_title('ax4')
x = np.random.randint(1, 101, 200)
y = np.random.randint(1, 101, 200)
ax4.scatter(x, y) # scatter と記述

plt.show() # グラフを表示させる為
```
